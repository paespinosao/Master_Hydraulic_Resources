```python
x=np.array([float(lcdata.x.values[i]) for i in range(0,len(lcdata.x.values))])
y=np.array([float(lcdata.y.values[i]) for i in range(0,len(lcdata.y.values))])
lc=np.column_stack((x,y))
print(x.shape,y.shape,lc.shape)
```
